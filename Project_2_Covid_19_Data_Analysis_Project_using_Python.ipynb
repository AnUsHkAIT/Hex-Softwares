```python
# Install necessary libraries
!pip install pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
confirmed_cases = pd.read_csv(url)


recovered_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
recovered_cases = pd.read_csv(recovered_url)

fatalities_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
fatalities_cases = pd.read_csv(fatalities_url)
```

```python
print(confirmed_cases.head())

print(confirmed_cases.info())

print(confirmed_cases.isnull().sum())
```

```python
confirmed_cases = confirmed_cases.groupby('Country/Region').sum().iloc[:, 1:]

print(confirmed_cases.head())


confirmed_cases = confirmed_cases.transpose()
print(confirmed_cases.head())
```

```python
plt.figure(figsize=(14, 7))
plt.plot(confirmed_cases['India'], label='Confirmed Cases in India', color='blue')
plt.title('Covid-19 Confirmed Cases in India Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()


plt.figure(figsize=(14, 7))
plt.plot(confirmed_cases.sum(axis=1), label='Global Confirmed Cases', color='red')
plt.title('Global Covid-19 Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
vaccination_url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'
vaccination_data = pd.read_csv(vaccination_url)


print(vaccination_data.head())
```

```python

correlation = vaccination_data['total_vaccinations'].corr(confirmed_cases.sum(axis=1))
print('Correlation between total vaccinations and confirmed cases:', correlation)
```
