```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert to a DataFrame for easier exploration
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({i: name for i, name in enumerate(iris.target_names)})

print(iris_df.head())
```

```python
print(iris_df.info())

print(iris_df.describe())

print("Missing values:\n", iris_df.isnull().sum())
```

```python
# Plot histograms for each feature
iris_df[iris.feature_names].hist(figsize=(12, 10), bins=20)
plt.suptitle('Distribution of Each Feature')
plt.show()
```

```python
# Pairplot to visualize relationships between features and species
sns.pairplot(iris_df, hue='species', palette='Set1')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Boxplots for each feature by species
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.boxplot(data=iris_df, x='species', y='sepal length (cm)', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length by Species')

sns.boxplot(data=iris_df, x='species', y='sepal width (cm)', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width by Species')

sns.boxplot(data=iris_df, x='species', y='petal length (cm)', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length by Species')

sns.boxplot(data=iris_df, x='species', y='petal width (cm)', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```
